## Задание 1

Будем парсить страницу со свежеми новостям на habr.com/ru/all/.
Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:
KEYWORDS = ['python', 'парсинг']
Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).
В итоге должен формироваться датафрейм со столбцами: <дата> - <заголовок> - <ссылка>.

```python
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
KEYWORDS = ['python', 'парсинг'] 
```

```python
all_news = pd.DataFrame()

for page in range(50):
    time.sleep(0.2)
    soup = BeautifulSoup(requests.get(f'https://habr.com/ru/all/page{page+1}/').text, 'html.parser')
    for post in soup.find_all('article', class_='post post_preview'):
        if any([kwd in str(post).lower() for kwd in KEYWORDS]):
            all_news = (
                pd.concat([all_news,
                pd.DataFrame([
                {'page':str(page+1),
                'date': post.find('span', class_='post__time').text,
                'title': post.find('a', class_='post__title_link').text,
                'url': post.find('a', class_='post__title_link').attrs.get('href')}])],
                ignore_index=True)
                    )
all_news
```

## Задание 2

Написать скрипт, который будет проверять список e-mail адресов на утечку при помощи сервиса Avast Hack Ckeck. Список email-ов задаем переменной в начале кода: EMAIL = [xxx@x.ru, yyy@y.com]
В итоге должен формироваться датафрейм со столбцами: - <дата утечки> - <источник утечки> - <описание утечки>.

```python
import requests
import json
import pandas as pd
```

```python
def get_email_leaks(email):
    res = []
    URL = 'https://digibody.avast.com/v1/web/leaks'
    params = {'email': email}
    resp = requests.post(URL, data=json.dumps(params))
    data = json.loads(resp.text)
    for leak in data['value']:
        res.append({
            'email': email,
            'date': leak['leak_info']['date'],
            'source': leak['leak_info']['title'],
            'description': leak['leak_info']['description']
        })
    return res

emails = ['xxx@x.ru', 'yyy@y.com']
data = []
for email in emails:
    data.extend(get_email_leaks(email))
```

```python
leaks_df = pd.DataFrame(data)
leaks_df['date'] = pd.to_datetime(leaks_df['date'], unit='ms')
leaks_df = leaks_df.sort_values(by='date', ascending=False)
leaks_df
```
